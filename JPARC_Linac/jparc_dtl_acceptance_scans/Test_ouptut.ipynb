```python
plt.plot(df_jparc.position, df_jparc.betaX)
plt.title('JPARC Linac Output');
plt.grid(ls=':');
plt.xlabel('s [m]');
plt.ylabel(r'$\beta_x$ [m]');
```

```python
plt.plot(df_jparc.position, df_jparc.betaY)
plt.title('JPARC Linac Output');
plt.grid(ls=':');
plt.xlabel('s [m]');
plt.ylabel(r'$\beta_y$ [m]');
```

```python
plt.plot(df_jparc.position, df_jparc.emittX)
plt.title('JPARC Linac Output');
plt.grid(ls=':');
plt.xlabel('s [m]');
plt.ylabel(r'$\epsilon_x$ [mm mrad?]');
```

```python
plt.plot(df_jparc.position, df_jparc.normEmittX)
plt.title('JPARC Linac Output');
plt.grid(ls=':');
plt.xlabel('s [m]');
plt.ylabel(r'$\frac{\epsilon_x}{\beta \gamma}$ [mm mrad?]');
```

```python
plt.plot(df_jparc.position, df_jparc.emittY)
plt.title('JPARC Linac Output');
plt.grid(ls=':');
plt.xlabel('s [m]');
plt.ylabel(r'$\epsilon_y$ [mm mrad?]');
```
